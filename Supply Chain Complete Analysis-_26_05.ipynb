### 3.1 Data Cleaning — Column Removal Rationale

The raw dataset has 53 columns. Not all of them are needed for delivery performance and profitability analysis. Columns were removed based on three criteria:

| Removal Reason | Columns | Justification |
|---|---|---|
| **PII / Privacy** | Customer Email, Customer Password, Customer Fname, Customer Lname, Customer Street | Personally Identifiable Information — must be excluded from analytical datasets per data privacy best practices |
| **100% Missing** | Product Description, Product Image | Entirely null — no analytical value |
| **High Missingness** | Order Zipcode (86% missing), Customer Zipcode | Too sparse to derive reliable insights; city/region columns provide sufficient geographic granularity |
| **Redundant / Duplicate** | Benefit per Order (≡ Order Profit Per Order), Category Id + Department Id (names already present), Product Category Id, Order Item Cardprod Id | Exact duplicates or ID columns where string names are more readable |
| **Zero Variance** | Product Status (single value across all rows) | A constant column adds no discriminatory information to any model or groupby |
| **Granularity Reduction** | Customer City, Order City, Customer State, Order State, Order Country | Geographic analysis is conducted at Region and Market level; city/state columns add noise without proportionate insight |

After cleaning: **172,765 rows × 20 columns** — all remaining columns are analytically relevant.

### 3.2 Feature Engineering

The raw dataset contains order dates and shipping dates, but not derived time and delay metrics. We create the following engineered features:

| Feature | Formula | Purpose |
|---|---|---|
| `Order Processing Time` | shipping_date − order_date (days) | Actual lead time from order to dispatch |
| `Delay` | Processing Time − Scheduled Days | Signed delay: positive = late, negative = early |
| `Is_Delayed` | Delay > 0 | Boolean flag for delay classification |
| `order_month` | month of order date | Enables seasonality analysis |
| `order_day` | day name of order date | Enables day-of-week pattern analysis |
| `order_hour` | hour of order date | Enables intra-day order timing analysis |
| `Profitability Flag` | Profit/Loss/Break-even | Categorises order profitability for segmentation |

```python
#calculating order processing time and delay
df ['Order Processing Time'] = (df['shipping date (DateOrders)'] - df['order date (DateOrders)']).dt.days
df ['Delay'] = df['Order Processing Time'] - df ['Days for shipment (scheduled)']
df ['Is_Delayed'] = df ['Delay'] > 0
df ['order_month'] = df ['order date (DateOrders)'].dt.month
df ['order_day'] =df['order date (DateOrders)'].dt.day_name()
df ['order_hour'] = df['order date (DateOrders)'].dt.hour
df.describe()
```


##  Bottleneck Detection

Here we identify *where* delays are concentrated — across shipping modes, customer segments, regions, and product departments.

The `compute_delay_pct_by_category()` function below is a reusable helper that computes late delivery percentage for any categorical dimension. This avoids repeating the same aggregation logic across multiple cells.


```python
# ── Reusable delay percentage calculator ────────────────────────────────────
# Accepts any categorical column name and returns:
#   - Total orders per category
#   - Late orders per category
#   - Late delivery % (sorted descending, top 10)
# Used across Shipping Mode, Customer Segment, Region, and Department analyses


def compute_delay_pct_by_category(category):
    
    cat_df = df.groupby(category).agg(
        total_orders=('Delay', 'count'),
        late_orders=('Is_Delayed', 'sum')
    ).reset_index()

    cat_df['delay_pct'] = (
        cat_df['late_orders']
        / cat_df['total_orders']
        * 100
    )

    cat_df = cat_df.sort_values(
        'delay_pct',
        ascending=False
    ).head(10)

    return cat_df


categories = [
    'Order Region',
    'Customer Segment',
    'Shipping Mode',
    'Order Status',
    'Type',
    'Department Name'
]

fig, axes = plt.subplots(
    2,
    3,
    figsize=(16, 7),
    constrained_layout=True
)

axes = axes.flatten()

for ax, category in zip(axes, categories):

    cat_df = compute_delay_pct_by_category(category)

    sns.barplot(
        data=cat_df,
        x='delay_pct',
        y=category,
        ax=ax,
        palette='viridis'
    )

    ax.set_title(f'Delay % by {category}')
    ax.set_xlabel('')
    ax.set_ylabel(category)

    for i, row in cat_df.reset_index().iterrows():

        ax.text(
            row['delay_pct'] - 15,
            i,
            f"{row['delay_pct']:.1f}%",
            va='center',
            fontsize=10,
            color='white'
        )

plt.show()
```


## Root Cause Analysis

Bottleneck detection showed *where* delays occur. Root cause analysis asks *why*  digging into the relationship between shipping mode, scheduled times, processing times, and delivery outcomes.

**Hypothesis:** Standard Class shipping is likely being applied to orders with tight delivery windows, creating a structural mismatch between promise and fulfilment capacity.

```python
# Top drivers of late delivery by region

def top_drivers_for_region(region):

    # Filter region data
    df_region = df[df['Order Region'] == region].copy()

    # Driver columns
    drivers = [
        'Shipping Mode',
        'Customer Segment',
        'Department Name',
        'Type',
        'Order Status'
    ]

    all_factors = []

    # Analyze each driver
    for factor in drivers:

        temp = (
            df_region.groupby(factor)
            .agg(
                total_orders=('Delay', 'count'),
                late_orders=('Is_Delayed', 'sum'),
                avg_delay=('Delay', 'mean')
            )
            .reset_index()
        )

        # Calculate delay percentage
        temp['delay_pct'] = (
            temp['late_orders'] / temp['total_orders']
        ) * 100

        # Add metadata columns
        temp['Driver'] = factor

        temp['Factor_Level'] = (
            factor + " : " + temp[factor].astype(str)
        )

        # Keep useful columns
        all_factors.append(
            temp[
                [
                    'Driver',
                    'Factor_Level',
                    'delay_pct',
                    'avg_delay',
                    'total_orders'
                ]
            ]
        )

    # Combine all driver data
    final_df = pd.concat(all_factors)

    # Top 10 delay drivers
    top_factors = (
        final_df
        .sort_values('delay_pct', ascending=False)
        .head(10)
    )

    # Plot
    plt.figure(figsize=(12, 6))

    bars = plt.barh(
        top_factors['Factor_Level'],
        top_factors['delay_pct'],
        color='crimson'
    )

    plt.xlabel("Delay Percentage (%)")

    plt.ylabel("Driver Factors")

    plt.title(f"Top Drivers of Late Delivery in {region}")

    plt.grid(True, linestyle=':', alpha=0.5)

    plt.gca().invert_yaxis()

    # Add labels on bars
    for bar in bars:

        width = bar.get_width()

        plt.text(
            width - 10,
            bar.get_y() + bar.get_height() / 2,
            f"{width:.1f}%",
            va='center',
            fontsize=10,
            color='white'
        )

    plt.show()


# Run function
top_drivers_for_region('Central Africa')
```

## Time Based Analysis

Operational bottlenecks often have a temporal signature — certain months, days, or hours that show disproportionately high delay rates. This section looks for those patterns.

If delays spike consistently in specific months, it points to seasonal demand surges or capacity constraints. If spikes occur on specific days or hours, it may indicate staffing or shift-related processing bottlenecks.

```python
# Delay % by Month, Day of Week, Hour
# Delay % by Month
delay_by_month = (
    df.groupby('order_month')['Is_Delayed']
    .mean()
    .reset_index()
)

delay_by_month['delay_pct'] = (
    delay_by_month['Is_Delayed'] * 100
)


# Delay % by Day of Week
delay_by_day = (
    df.groupby('order_day')['Is_Delayed']
    .mean()
    .reset_index()
)

delay_by_day['delay_pct'] = (
    delay_by_day['Is_Delayed'] * 100
)


# Delay % by Hour
delay_by_hour = (
    df.groupby('order_hour')['Is_Delayed']
    .mean()
    .reset_index()
)

delay_by_hour['delay_pct'] = (
    delay_by_hour['Is_Delayed'] * 100
)
```

```python
# Balancing the training data using SMOTE
print("Before balancing (train):", Counter(y_train))
smote = SMOTE(random_state=42)
X_train_bal, y_train_bal = smote.fit_resample(X_train, y_train)
print("After balancing (train):", Counter(y_train_bal))
```

```python
def evaluate_model(y_true, y_pred, model_name):
  print(f"\n--- model_name ---")
  print("Accuracy:", round(accuracy_score(y_true, y_pred),2))
  print("Precision:", round(precision_score(y_true, y_pred),2))
  print("Recall:", round(recall_score(y_true, y_pred),2))
  print("\nClassification Report:\n", classification_report(y_true, y_pred))
```

```python
# Fit the Random Forest model on balanced data
rf_model_balanced = RandomForestClassifier(random_state=42)
rf_model_balanced.fit(X_train_bal, y_train_bal)
y_pred_rf_balanced = rf_model_balanced.predict(X_test)
evaluate_model(y_test, y_pred_rf_balanced, "Random Forest Classifier")
```
